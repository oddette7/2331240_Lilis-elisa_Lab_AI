```python
import sys
import importlib

print("Python version:", sys.version)

def verify_package(package_name):
    try:
        module = importlib.import_module(package_name)
        version = getattr(module, '__version__', 'unknown')
        print(f"'{package_name}' package: OK, version = {version}")
    except ImportError:
        print(f"'{package_name}' package: ERROR - not found")

print("---")
verify_package('torch')
verify_package('tensorflow')
print("---")
```
